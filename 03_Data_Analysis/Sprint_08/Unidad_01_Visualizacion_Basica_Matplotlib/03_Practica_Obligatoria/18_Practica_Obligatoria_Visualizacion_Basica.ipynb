#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Top 5 por poblacin
top5 = df.sort_values("population_total", ascending=False).head(5)

fig1, ax = plt.subplots(figsize=(8, 6))

colors = ["blue", "grey", "olive", "saddlebrown", "peru"]

bars = ax.bar(
    top5["city"],
    top5["population_total"] / 1000,        # en miles
    color=colors,
    alpha=0.7
)

ax.set_title("Ciudades ms pobladas de California")
ax.set_ylabel("Habitantes (miles)")

# Etiquetas encima de cada barra
for bar, value in zip(bars, top5["population_total"] / 1000):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 50,
        f"{value:.3f}",
        ha="center",
        va="bottom"
    )

plt.show()
```

```python
fig2, ax = plt.subplots(figsize=(8, 6))

ax.hist(
    df["area_total_km2"], 
    bins=50, 
    color="steelblue", 
    edgecolor="white", 
    label="Hist Area"
)

ax.legend()
ax.set_ylabel("Nmero de ciudades")

# Etiqueta del eje X con la fuente, tamao y color indicados
ax.set_xlabel(
    "Superficie ($km^2$)",
    fontfamily="Bradley Hand ITC",
    fontsize=18,
    color="steelblue"
)

plt.show()
```

```python
fig3, ax = plt.subplots(figsize=(8, 6))

ax.scatter(
    df["area_water_km2"],
    df["population_total"] / 1000,
    color="green"
)

ax.set_xlabel("rea de agua en la ciudad ($km^2$)")
ax.set_ylabel("Poblacin (miles habitantes)")
ax.set_title("Poblacin por kilmetro cuadrado de agua")

```

```python
fig4, ax = plt.subplots(figsize=(8, 6))

ax.boxplot(
    df["elevation_m"],
    vert=True,                           # Cambiar a vertical
    patch_artist=True,
    boxprops=dict(facecolor="red", alpha=0.7)
)

ax.set_title("Diagrama de caja de la Altitud de las ciudades de California")
ax.set_ylabel("Altitud (m)")

```

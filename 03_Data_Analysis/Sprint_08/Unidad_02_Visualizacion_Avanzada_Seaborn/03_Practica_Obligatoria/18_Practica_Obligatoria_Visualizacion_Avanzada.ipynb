### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def pinta_dispersion(df, x, y, cat=None, tam=40, escala=1.0, modo="leyenda"):
    # (1)(2)(3) df + columnas numricas (X,Y)
    if x not in df.columns or y not in df.columns:
        raise ValueError("x e y deben existir en df.")
    if x == y:
        raise ValueError("x e y deben ser diferentes.")

    # (4) columna categrica opcional
    if cat is not None and cat not in df.columns:
        raise ValueError("cat debe existir en df o ser None.")

    # (5)(6) tamao: nmero o columna + escala
    tam_es_col = isinstance(tam, str)
    dfp = df.copy()
    col_size = None

    if tam_es_col:
        if tam not in dfp.columns:
            raise ValueError("Si tam es str, debe ser una columna de df.")
        col_size = "__size__"
        dfp[col_size] = dfp[tam].astype(float) * float(escala)
        v = dfp[col_size].to_numpy()
        v = np.nan_to_num(v, nan=0.0, posinf=0.0, neginf=0.0)
        v = np.clip(v, a_min=1e-6, a_max=None)
        dfp[col_size] = v

    plt.figure(figsize=(8, 5))

    # EXTRA: colorbar si no hay cat y tam es columna
    if modo == "colorbar" and cat is None and tam_es_col:
        sizes_raw = dfp[col_size].to_numpy()
        smin, smax = sizes_raw.min(), sizes_raw.max()
        if smax > smin:
            sizes = 20 + (sizes_raw - smin) * (300 - 20) / (smax - smin)
        else:
            sizes = np.full_like(sizes_raw, 60.0)

        sc = plt.scatter(dfp[x], dfp[y], s=sizes, c=sizes_raw)
        plt.colorbar(sc)
        plt.xlabel(x); plt.ylabel(y)
        plt.title(f"{y} vs {x}")
        plt.tight_layout()
        plt.show()
        return

    # cat=None -> NO pasar hue
    if cat is None:
        if tam_es_col:
            sns.scatterplot(data=dfp, x=x, y=y, size=col_size, sizes=(20, 300),
                            legend=("full" if modo == "leyenda" else True))
        else:
            sns.scatterplot(data=dfp, x=x, y=y, s=float(tam), legend=False)
    else:
        if tam_es_col:
            sns.scatterplot(data=dfp, x=x, y=y, hue=cat, size=col_size, sizes=(20, 300),
                            legend=("full" if modo == "leyenda" else True))
        else:
            sns.scatterplot(data=dfp, x=x, y=y, hue=cat, s=float(tam),
                            legend=("full" if modo == "leyenda" else True))

    plt.title(f"{y} vs {x}")
    plt.tight_layout()
    plt.show()


```

### #1.3

Pinta el diagrama de dispersin de longitud y latitud (longitud en las x, latitud en las y) de las ciudades, usando el tamao de las ciudades para el tamao de los puntos y los colores para mostrar la poblacin. Previamente crea una columna nueva "log_poblacion" y asignale el log decimal de la poblacion (np.log10() y como argumento los valores de la columna "population_total"). Usa "log_poblacion" para el color.

```python
# (1) Nueva columna log_poblacion
df["log_poblacion"] = np.log10(df["population_total"])

# (2) Tamao de puntos = tamao de la ciudad (usa el rea; ajusta la escala si lo ves muy grande/pequeo)
sizes = df["area_total_km2"].astype(float) * 5  # escala

# (3) Dispersin: longd (X) vs latd (Y), color = log_poblacion
plt.figure(figsize=(10, 5))
sc = plt.scatter(
    df["longd"], df["latd"],
    s=sizes,
    c=df["log_poblacion"],
    alpha=0.7
)
plt.colorbar(sc, label="log10(population_total)")
plt.xlabel("longd")
plt.ylabel("latd")
plt.title("Ciudades de California: longitud vs latitud")
plt.tight_layout()
plt.show()
```

Tienes que entregar ambas cosas: presentacin y notebook :)

```python
# Cargar datos
df = sns.load_dataset("titanic")

# Limpieza mnima
df = df.dropna(subset=["sex", "age", "class", "embarked", "survived"])

# Crear grupo de edad (nio / adulto)
df["age_group"] = np.where(df["age"] < 18, "child", "adult")
```

```python
plt.figure(figsize=(5,4))
df["survived"].value_counts(normalize=True).plot(kind="bar")
plt.ylabel("Proporcin")
plt.xlabel("Survived (0 = No, 1 = S)")
plt.title("Alta tasa de mortalidad en el Titanic")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))
df.groupby("sex")["survived"].mean().plot(kind="bar")
plt.ylabel("Tasa de supervivencia")
plt.title("Supervivencia por sexo")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))
df.groupby("class")["survived"].mean().plot(kind="bar")
plt.ylabel("Tasa de supervivencia")
plt.title("Supervivencia por clase")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="sex", y="survived", hue="age_group")
plt.ylabel("Tasa de supervivencia")
plt.title("Supervivencia: mujeres vs nios")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))
df.groupby("embarked")["survived"].mean().plot(kind="bar")
plt.ylabel("Tasa de supervivencia")
plt.title("Supervivencia por puerto de embarque")
plt.tight_layout()
plt.show()
```
